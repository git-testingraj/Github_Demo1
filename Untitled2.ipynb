```python
def decor_wish(func):
    def decor_inner(name):
        if name == 'Raj':
            print('Hello Raj')
        else:
            func(name)
    return decor_inner
```

```python
@decor_wish
def wish(name):
    print(f'Hello {name},How r you ?')
```

```python
wish('Raj')
```
